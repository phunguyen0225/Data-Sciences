```python
#Data pre-processing
```

```python
cars_df['fueltype'] = cars_df['fueltype'].map({'gas':1, 'diesel':0})
cars_df['aspiration'] = cars_df['aspiration'].map({'std':1, 'turbo':0})
cars_df['doornumber'] = cars_df['doornumber'].map({'two':1, 'four':0})
cars_df['enginelocation'] = cars_df['enginelocation'].map({"front":1, "rear":0})
```

```python
cars_df.head()
```

```python
cars_df.drop(['car_ID'], inplace=True, axis=1)
```

```python
cars_df.head()
```

```python
body_dum = pd.get_dummies(cars_df['carbody'])
body_dum = pd.get_dummies(cars_df['carbody'], drop_first = True)
cars_df = pd.concat([cars_df, body_dum], axis = 1)
cars_df.drop(['carbody'], axis = 1, inplace=True)
cars_df.head()
```

```python
wheel_dum = pd.get_dummies(cars_df['drivewheel'])
wheel_dum = pd.get_dummies(cars_df['drivewheel'], drop_first=True)
cars_df = pd.concat([cars_df, wheel_dum], axis = 1)
cars_df.drop(['drivewheel'], axis = 1, inplace=True)
cars_df.head()
```

```python
engine_dum = pd.get_dummies(cars_df['enginetype'])
engine_dum = pd.get_dummies(cars_df['enginetype'], drop_first = True)
cars_df = pd.concat([cars_df, engine_dum], axis = 1)
cars_df.drop(['enginetype'], axis = 1, inplace=True)
cars_df.head()
```

```python
cylinder_dum = pd.get_dummies(cars_df['cylindernumber'])
cylinder_dum = pd.get_dummies(cars_df['cylindernumber'], drop_first=True)
cars_df = pd.concat([cars_df, cylinder_dum], axis = 1)
cars_df.drop(['cylindernumber'], axis =1, inplace=True)
cars_df.head()
```

```python
fuel_dum = pd.get_dummies(cars_df['fuelsystem'])
fuel_dum = pd.get_dummies(cars_df['fuelsystem'], drop_first = True)
cars_df = pd.concat([cars_df, fuel_dum], axis = 1)
cars_df.drop(['fuelsystem'], axis=1, inplace=True)
cars_df.head()
```

```python
name_dum = pd.get_dummies(cars_df['Company'])
name_dum = pd.get_dummies(cars_df['Company'], drop_first=True)
cars_df = pd.concat([cars_df, name_dum], axis = 1)
cars_df.drop(['Company'], axis = 1, inplace=True)
cars_df.head()
```

```python
from sklearn.model_selection import train_test_split
cars_train, cars_test = train_test_split(cars_df, train_size = 0.7, test_size = 0.3)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
var = ['symboling', 'carlength', 'carheight', 'enginesize', 'boreratio', 'stroke',
       'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'price', 'carwidth',
       'curbweight', 'highwaympg', 'wheelbase']
cars_train[var] = scaler.fit_transform(cars_train[var])
cars_train.head()
```

```python
#PART 6: Model Creation and Evaluation
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
y_train = cars_train.pop('price')
x_train = cars_train
```

```python
from sklearn.feature_selection import RFE
lm = LinearRegression()
lm.fit(x_train, y_train)

rfe = RFE(lm, 15)
rfe = rfe.fit(x_train, y_train)
```

```python
zip(x_train.columns, rfe.support_, rfe.ranking_)
```

```python
list(zip(x_train.columns, rfe.support_, rfe.ranking_))
```

```python
x_train.columns[~rfe.support_]
```

```python
col = x_train.columns[rfe.support_]
col
x_test = x_train[col]
```

```python
import statsmodels.api as sm
x_test = sm.add_constant(x_test)

lm = sm.OLS(y_train, x_test).fit()

print(lm.summary())
```

```python
print("The R-squared is 0.947\n")
```
