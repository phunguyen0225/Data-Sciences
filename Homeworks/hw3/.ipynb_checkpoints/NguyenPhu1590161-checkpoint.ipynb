```python
"""Problem Statement
A business provides you with the following data and wants you to find patterns in their customer's 
habits so that they can make better recommendations and target certain groups in the future.
Would we approach this as a supervised or unsupervised learning task and why?
"""
```

```python
#Answer: We would approach this as a unsupervised learning task because we are using k-means which is unsupervised machine learning
```

```python
#STEP 1
import numpy as np
import pandas as pd

data = pd.read_csv('hw3_data.csv')
print("Total observations: ", len(data.index))
print()
print("Number of missing values:\n", data.isnull().sum())
print()
print("Unique values in each column:\n", data.nunique())
```

```python
#STEP 2 (VISUALIZATIONS)
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['figure.figsize'] = 15, 5
import warnings
warnings.filterwarnings('ignore')

fig, axes = plt.subplots(nrows=1, ncols = 3)

sns.distplot(data['Age'], ax=axes[0])
axes[0].set_title('Age Distribution')

sns.distplot(data['Annual Income (k$)'], ax=axes[1])
axes[1].set_title('Annual Income (k$) Distribution')

sns.distplot(data['Spending Score (1-100)'], ax=axes[2])
axes[2].set_title('Spending Score (1-100)')
```

```python
#What i can conclude is that for the Age Distribution have a skewed-left distribution,
#so the mean is less than the median

#Annual Income and Spending Score is a shape that is almost but not exact to a bell curve
#so i would say the distribution is close to normal
```

```python
#VISUALIZATIONS CONTINUED
male = data['Gender'].value_counts()['Male']
female = data['Gender'].value_counts()['Female']
count_data = [male, female]
gender = ['Male', 'Female']
wp = { 'linewidth' : 2, 'edgecolor' : "black" }
plt.pie(count_data, labels = gender, shadow = True, wedgeprops = wp, startangle = 200)
plt.show()
```

```python
#What i can conclude is that there are more number of female than male in the data
```

```python
#VISUALIZATIONS CONTINUED
sns.heatmap(data.corr(), annot=True, cmap="YlGnBu", linewidths=.1)
```

```python
#There is significant collinearity between CustomerID and Annual Income because the correlation is < 0.8
```

```python
#VISUALIZATION CONTINUED
fig, axes = plt.subplots(nrows=1, ncols = 3)

sns.regplot(data['Age'], data['Annual Income (k$)'], ax=axes[0])
axes[0].set_title('Age vs. Annual Income (k$)')

sns.regplot(data['Age'], data['Spending Score (1-100)'], ax=axes[1])
axes[1].set_title('Age vs. Spending Score (1-100)')

sns.regplot(data['Annual Income (k$)'], data['Spending Score (1-100)'], ax=axes[2])
axes[2].set_title('Annual Income (k$) vs. Spending Score (1-100)')
```

```python
#What i can conclude is that for Age vs. Annual Income and Annual Income vs Spending Score is a constant linear
#you can say as Age increases the number of annual incomr increases, also the same for annual income vs spending score

#Age vs. Spending Score (1-100) have a negative linear, as age increases the number of spending score (1-100) decreases
```

```python
#VISUALIZATION CONTINUED

sns.lineplot(x=data['Annual Income (k$)'], y=data['Age'])
sns.lineplot(x=data['Annual Income (k$)'], y = data['Spending Score (1-100)'])
```

```python
#As annual income increases, the age and spending score fluctuate almost parallel but number is age is lower
```

```python
#STEP 3 (k-means clustering)
df = pd.DataFrame({'spendingScore': data['Spending Score (1-100)'], 'annualIncome': data['Annual Income (k$)']})

from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
model = KMeans()
visualizer = KElbowVisualizer(model, k = (2, 30), timings = True)
visualizer.fit(df)
visualizer.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

SpendingAnnual = []

for i in range(2,11):
    SpendingAnnual.append(silhouette_score(df, AgglomerativeClustering(n_clusters = i).fit_predict(df)))
    
plt.bar(x=range(2,11), height=SpendingAnnual)
plt.xlabel('Number of Clusters')
plt.ylabel('Spending Score and Annual Income')
plt.title("Scores")
plt.show()
    
```

```python
#STEP 4 (hierarchical Clustering)
import scipy.cluster.hierarchy as shc

plt.title('Visualizing Data')
Dendrogram = shc.dendrogram((shc.linkage(df, method = 'ward')))
```

```python
#STEP 5

#In K-means clustering, it use a prespecifice number of clusters, the method assigns record to each cluster to find the 
#mutually exclusive cluster of spherical shape based on distace. Kmeans clustering needed advance knowledge of K number of 
#cluster one want to divide the data.

#In Hierarchical clustering, the methods can be either divisive or agglomerative. In hierarchical clustering, one can stop
#at any number of clusters. Hierarchical clustering requires the computation and storage of an nxn distance matrix. For 
#very large datasets, this can be expensive and slow. 

#The advantages of using Kmeans is covergence is guaranteed. Specialized to clusters of different sizes and shape. 
#The disadvatage of using kmeans is K-value is difficult to predict. Didn't work well with global cluster

#The advantage of using Hierarchy clustering is ease of handling of any forms of similarity or distance. Consequently,
#applicability to any attributes type. The disadvantage is computation cost.

#For me the hierarchy clustering take more time then the kmeans clustering


#I would use KMeans if there is a specific number of clusters in the dataset, but the group they belong to is unknown. Also,
#any data with a large number of variables because the computation time of hierarchical is expensive

#I would use Hierarchical clustering if the number of variables is low, if the distinguishes are based on prior beliefts,
# hierarchical clustering should be used to know the number of cluster because it is easier to determine the number of
#cluster with the dendrogram
```
